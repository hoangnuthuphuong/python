**[ Cấu trúc: numpy. apply_along_axis ( func1d , axis , arr , * args , * kwargs ) ]**

=>Là hàm cho phép ta sử dụng hàm theo từng chiều.

* hàm func1d (M,) -> (Nj…) : Hàm thực hiện
* axis : số nguyên trục
* Trục dọc theo arr được cắt. arr :ndarray (Ni…, M, Nk…)
* Đầu vào mảng. args: bất kỳ
* Đối số bổ sung cho func1d . kwargs: bất kỳ
*args và *kwargs đều chủ yếu được sử dụng trong định nghĩa. Hai cú pháp đặc biệt này giúp chúng ta có thể truyền bao nhiêu tham số vào hàm cũng được. Hai hàm này thông thường ta thường bỏ qua vì trong điều kiện bình thường không cần sử dụng.

```python
import numpy as np
```

```python
A = np.random.randint(1,20,[15,17])
print("Ma trận A:\n",A)
```

```python
vector_a =np.array(A)
def tim_gia_tri_lon_nhat(vector_a):
  return np.max(vector_a)
np.apply_along_axis(tim_gia_tri_lon_nhat,1,vector_a)


  
```
