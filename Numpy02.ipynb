```python
#Nhân hai ma trận
A = np.random.randint(1,5,[5,7])
print("Ma trận A:\n", A)
B = np.random.randint(1,5,[7,3])
print("Ma trận B:\n", B)
print("Nhân hai ma trận A và B:\n",A.dot(B))
#Ngoài ra: A @ B cũng là phép nhân hai ma trận A và B
```
