**1. numpy.arange()**

Hm ny tr v mt i tng ndarray cha cc gi tr cch u nhau trong mt phm vi nht nh. C php nh sau :

>**numpy.arange(start, stop, step, dtype)**

* start : Khong thi gian bt u. Nu b qua, mc nh l 0
* stop : Kt thc khong thi gian 
* step : Khong cch gia cc gi tr, mc nh l 1
* dtype : Kiu d liu ca ndarray kt qu. Nu khng c cung cp, kiu d liu u vo s c s dng

```python
a = np.arange(1,8)
print(a)
```

**2. numpy.linspace()**

Hm ny tng t nh hm arange (). Trong hm ny, thay v kch thc bc, s lng cc gi tr cch u gia khong c ch nh. 

>**numpy.linspace(start, stop, num, endpoint, retstep, dtype)**

Lu  cc tham s sau :


*   start : Khong thi gian bt u. Nu b qua, mc nh l 0

*   stop : Kt thc khong thi gian 

*   num : S lng mu cch u nhau s c to ra. Mc nh l 50

*   endpoint :Mc nh l true, do  gi tr dng c bao gm trong chui. Nu sai, n khng c gi tr dng

*   retstep : Nu ng, tr v cc gi tr mu v bc gia cc s lin tip

*   dtype : tr v kiu d liu l ndarray

```python
a = np.linspace(10,20,5) 
print (a)

a = np.linspace(10,20, 8, endpoint = False)
print(a)

a = np.linspace(10,20,5, retstep = True) 
print (a)

```

**3. numpy.logspace**

Hm ny tr v mt i tng ndarray c cha cc s cch u nhau trn thang log. im cui bt u v im dng ca thang o l ch s ca c s, thng l 10.

> **numpy.logspace(start, stop, num, endpoint, base, dtype)**


* start : Khong thi gian bt u. Nu b qua, mc nh l 0
* stop : Kt thc khong thi gian 
* num : S lng mu cch u nhau s c to ra. Mc nh l 50
* endpoint :Mc nh l true, do  gi tr dng c bao gm trong chui. Nu sai, n khng c gi tr dng
* base : C s ca khng gian nht k, mc nh l 10
* dtype : tr v kiu d liu l ndarray

```python
a = np.logspace(1, 2, num = 8) 
print (a)

a = np.logspace(1,10,num = 8, base = 2) 
print (a)
```

Hm tnh gi tr trung bnh: **np.mean()**

```python
x = np.array([1,2,3,4,5])
print("Gi tr trung bnh= ",np.mean(x))
```

Hm tnh gi tr trung v: **np.median()**

```python
x = np.array([1,2,3,4,5,6])
print("Gi tr trung v= ",np.median(x))
```

Hm tnh gi tr yu v : **np.nanmedian()**

```python
x = np.array([1,2,3,4,5,6,7])
print("Gi tr yu v= ",np.nanmedian(x))
```

Hm tnh phng sai: **np.var()**

```python
x = np.array([1,2,3,4,6,7,8])
print("Phng sai= ",np.var(x))
```

Hm tnh  lch chun: **np.std()**

```python
print(" lch chun= ", np.std(x))
```

```python
print("Phng sai tng hng ca a = ",np.var(a,axis=1))
print("Phng sai tng ct ca a = ",np.var(a,axis=0))
```

```python
print(" lch chun tng hng ca a = ",np.std(a,axis=1))
print(" lch chun tng ct ca a = ",np.std(a,axis=0))
```
