#**Bài 1: Excavation Depth and Archaeology-Độ sâu khai quật và khảo cổ học**

Bốn địa điểm khai quật khác nhau tại một khu vực khảo cổ ở New Mexico đã đưa ra độ sâu sau (cm) cho những khám phá khảo cổ học quan trọng.

* X1 = độ sâu tại Vị trí I

* X2 = độ sâu tại Vị trí II

* X3 = độ sâu tại Vị trí III

* X4 = độ sâu tại Vị trí IV 

**Xây dựng giả thuyết :**

$H_{0}:$ Không có sự khác nhau về độ sâu trung bình của bốn địa điểm khai quật

$H_{a}:$ Có ít nhất một địa điểm có độ sâu trung bình khác với các địa điểm còn lại

```python
df1 = pd.read_excel("owan01 (1).xls", engine = "xlrd")
df1
```

```python
#Loại bỏ các giá trị lỗi
df1 = df1.dropna()
df1
```

**Kiểm tra mẫu có tuân theo phân phối chuẩn**

Phát biểu giả thuyết :

Giả thuyết null

$$H_0 : \text{Các nhóm được lấy từ phân phối chuẩn}$$

Giả thuyết thay thế 

$$H_a:\text{Các nhóm không được lấy từ phân phối chuẩn}$$ 

```python
df1_new= [list(df1["X1"]), list(df1["X2"]), list(df1["X3"]), list(df1["X4"])]
```

```python
for i in range(4):
  statistic, p_value = st.shapiro(df1_new[i])
  print(f"P_value của nhóm X{i+1}: {p_value}")
  if p_value < 0.05:
    print(f"Có bằng chứng bác bỏ giả thuyết nhóm X{i+1} lấy từ phân phối chuẩn")
  else:
    print(f"Không có bằng chứng bác bỏ giả thuyết nhóm X{i+1} lấy từ phân phối chuẩn")
```

**Kiểm tra sự khác nhau về phương sai**

Phát biểu giả thuyết :

Giả thuyết null

$$H_0 : \text{Các nhóm có phương sai giống nhau}$$

Giả thuyết thay thế 

$$H_a:\text{Các nhóm có phương sai không giống nhau}$$ 

```python
stats, p_value = st.bartlett(df1["X1"], df1["X2"], df1["X3"], df1["X4"])
print("p_value:",p_value)
if p_value < 0.05:
  print("Có bằng chứng bác bỏ giả thuyết phương sai các nhóm bằng nhau")
else:
  print("Không có bằng chứng bác bỏ giả thuyết phương sai các nhóm bằng nhau")
```

✈ Mẫu được lấy từ phân phối chuẩn 

✈ Không có sự khác biệt về phương sai giữa các nhóm

**✈✈✈ TIẾN HÀNH KIỂM ĐỊNH ANOVA**

Phát biểu giả thuyết:

Giả thuyết null

$$H_0 : \text{Độ sâu trung bình của 4 địa điểm khai quật không có sự khác biệt}$$

Giả thuyết thay thế 

$$H_a:\text{Tồn tại ít nhất sự khác biệt giữa Độ sâu trung bình của 4 địa điểm khai quật}$$

```python
stats, p_value = st.f_oneway(df1["X1"], df1["X2"], df1["X3"], df1["X4"])
print("p_value:",p_value)
if p_value < 0.05:
  print("Có bằng chứng bác bỏ giả thuyết H0")
else:
  print("Không có bằng chứng bác bỏ giả thuyết H0")
```

**Kết luận:**  Có sự khác nhau về độ sâu trung bình của 4 địa điểm khai quật

# **Apple Orchard Experiment-Thử nghiệm Apple Orchard**

Năm loại gốc rễ đã được sử dụng trong một thí nghiệm ghép vườn táo. Dữ liệu sau đây thể hiện mức tăng trưởng mở rộng (cm) sau bốn năm.

* X1 = tăng trưởng mở rộng cho loại I

* X2 = tăng trưởng mở rộng cho loại II

* X3 = tăng trưởng mở rộng cho loại III

* X4 = tăng trưởng mở rộng cho loại IV

* X5 = tăng trưởng mở rộng cho loại V

Xây dựng giả thuyết :
$$H_{0}: \mu_{1} = \mu_{2} = ... = \mu_{n}$$
$$H_{a}: \text{Tồn tại ít nhất một cặp } (\mu_{i}, \mu_{j}) \text{ có giá trị khác nhau}$$

```python
df2 = pd.read_excel("owan02.xls", engine = "xlrd")
df2
```

```python
#Loại bỏ các giá trị lỗi
df2 = df2.dropna()
df2
```

**Kiểm tra mẫu có tuân theo phân phối chuẩn**

Phát biểu giả thuyết :

Giả thuyết null

$$H_0 : \text{Các nhóm được lấy từ phân phối chuẩn}$$

Giả thuyết thay thế 

$$H_a:\text{Các nhóm không được lấy từ phân phối chuẩn}$$ 

```python
df2_new= [list(df2["X1"]), list(df2["X2"]), list(df2["X3"]), list(df2["X4"]), list(df2["X5"])]
```

```python
for i in range(5):
  statistic, p_value = st.shapiro(df2_new[i])
  print(f"P_value của nhóm X{i+1}: {p_value}")
  if p_value < 0.05:
    print(f"Có bằng chứng bác bỏ giả thuyết nhóm X{i+1} lấy từ phân phối chuẩn")
  else:
    print(f"Không có bằng chứng bác bỏ giả thuyết nhóm X{i+1} lấy từ phân phối chuẩn")
```

**Kiểm tra sự khác nhau về phương sai**

Phát biểu giả thuyết :

Giả thuyết null

$$H_0 : \text{Các nhóm có phương sai giống nhau}$$

Giả thuyết thay thế 

$$H_a:\text{Các nhóm có phương sai không giống nhau}$$ 

```python
stats, p_value = st.bartlett(df2["X1"], df2["X2"], df2["X3"], df2["X4"],df2["X5"])
print("p_value:",p_value)
if p_value < 0.05:
  print("Có bằng chứng bác bỏ giả thuyết phương sai các nhóm bằng nhau")
else:
  print("Không có bằng chứng bác bỏ giả thuyết phương sai các nhóm bằng nhau")
```

✈ Mẫu được lấy từ phân phối chuẩn 

✈ Không có sự khác biệt về phương sai giữa các nhóm

**✈✈✈ TIẾN HÀNH KIỂM ĐỊNH ANOVA**

Phát biểu giả thuyết:

Giả thuyết null

$$H_0 : \text{Tăng trưởng mở rộng của 5 loại không có sự khác biệt}$$

Giả thuyết thay thế 

$$H_a:\text{Tồn tại ít nhất sự khác biệt giữa Tăng trưởng mở rộng của 5 loại}$$

```python
stats,p_value = st.f_oneway(df2["X1"],df2["X2"],df2["X3"],df2["X4"],df2["X5"])
print("p_value:",p_value)
if p_value < 0.05:
  print("Có bằng chứng bác bỏ giả thuyết H0")
else:
  print("Không có bằng chứng bác bỏ giả thuyết H0")
```

**Kết luận:**  Tăng trưởng mở rộng của 5 loại gốc rễ không có sự khác biệt
